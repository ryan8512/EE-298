```python
import numpy as np
import matplotlib.pyplot as plt
```

<h1> 1. Functions needed for the program </h1>

<p> 1.1 get_angles function generate a 16x16 matrix which is distributed from 0 to 2pi. In other words, the 1st and 2nd columns are  from [0,1/4*pi); the 3rd and 4th column values is from [1/4*pi,2/4*pi)</p>

<p> 1.2 positional_encoding function simply add uses sin and mapping </p>

<p> 1.3 normalize_vector function normalizes vectors greater than 0 </p>

<p> 1.4 gram_schmidt function uses the gram_schmidt method to make the matrix orthogonal and the vectors orthonormal </p>


```python
# Code from https://www.tensorflow.org/tutorials/text/transformer
def get_angles(position,d_model):
    #Generate the positional vector
    pos = np.arange(position)[:, np.newaxis]
    pos = pos * 1/pos.shape[0]   #Normalize
    
    #Generate the dimensions vector
    dim = np.ones(d_model//2)*(1/4*np.pi)
    dim = np.repeat(dim,2)
    dim = dim[np.newaxis,:]
    
    output_matrix = pos * dim
    
    #Add the offset matrix
    offset_matrix = np.zeros([pos.shape[0],dim.shape[1]])
    for i in range(pos.shape[0]):
        #For interleaving the sin and cos: use np.repeat
        offset_matrix[i,:] = np.repeat( np.linspace(0, 2*np.pi,num=8,endpoint=False),2) 
    
    output_matrix = output_matrix + offset_matrix
    
    return output_matrix

def positional_encoding(position, d_model):
    #Output is a matrix which is distributed from 0 to 2pi across the matrix
    angle_rads = get_angles(position,d_model)
    
    #apply sin to even indices in the array; 2i
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
    #apply cos to odd indices in the array; 2i+1
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])
    
    #Redundant for now, but just wanted to change name
    pos_encoding = angle_rads
    return pos_encoding

def normalize_vector(input_vector):
    magnitude = np.linalg.norm(input_vector)
    return input_vector/magnitude

def gram_schmidt(input_matrix):
    cols = input_matrix.shape[1]
    rows = input_matrix.shape[0]
    output_matrix = np.zeros([rows,cols])
    for i in range(cols):
        buffer_vector = input_matrix[:,i]
        for j in range(i):
            buffer_vector = buffer_vector - np.dot(input_matrix[:,i],output_matrix[:,j])*output_matrix[:,j]
        output_matrix[:,i] = normalize_vector(buffer_vector)     ##Normalize
    return output_matrix

def is_orthogonal(input_matrix):
    for i in range(input_matrix.shape[1]):
        for j in range(input_matrix.shape[1]-i-1):
            if(np.round(np.dot(input_matrix[:,i],input_matrix[:,j+1]),decimals=2) != 0):
                return False
    return True

def is_normal(input_matrix):
    for i in range(input_matrix.shape[1]):
        if(np.round(np.linalg.norm(input_matrix[:,i]),decimals=2) != 1):
            return False
    return True
```

<h1> 2. Generate the 16x16 matrix </h1>

```python
#Parameters
tokens = 16
dimensions = 16
```

<h2> 2.1 Angles Generated </h2>

```python
#Angles (For Sanity Check)
print_angles = get_angles(tokens,dimensions)

#Print
plt.figure(figsize=(12,8))
plt.pcolormesh(print_angles, cmap='viridis')
plt.title('16x16 Angles')
plt.xlabel('Number of Columns')
plt.xlim((0, dimensions))
plt.ylim((tokens,0))
plt.ylabel('Number of Dimensions')
plt.colorbar()
plt.show()
```

<h2> 2.2 Positional Encoding </h2>

```python
#Apply sin and cos
pos_encoding = positional_encoding(tokens, dimensions)

#For printing
plt.figure(figsize=(12,8))
plt.pcolormesh(pos_encoding, cmap='viridis')
plt.title('16x16 positional encoding')
plt.xlabel('Number of Columns')
plt.xlim((0, dimensions))
plt.ylim((tokens,0))
plt.ylabel('Number of Dimensions')
plt.colorbar()
plt.show()
```

<h1> 3. Test for Orthogonality and Orthonormality </h1>

```python
print("Orthogonal: "+str(is_orthogonal(pos_encoding)))
print("Normal: "+str(is_normal(pos_encoding)))
```

<h1> 4. Apply G-S Method and Print </h1>

```python
orthonormal_matrix = gram_schmidt(pos_encoding)

plt.figure(figsize=(12,8))
plt.pcolormesh(orthonormal_matrix, cmap='viridis')
plt.title('G-S Method')
plt.xlabel('Number of Columns')
plt.xlim((0, dimensions))
plt.ylim((tokens,0))
plt.ylabel('Number of Dimensions')
plt.colorbar()
plt.show()
```

<h1> 5. Test for Orthogonality and Orthonormality AGAIN</h1>

```python
print("Orthogonal: "+str(is_orthogonal(orthonormal_matrix)))
print("Normal: "+str(is_normal(orthonormal_matrix)))
```

<h1> 6. Conclusion: Positional Vectors are not Linearly Independent when distributed as such </h1>
    <h2> Which makes sense! Given that from the angle graph, the third column and the sixteen columns are the same! </h2>
        <h3> In other words, sin(pi/4 - pi/2) == cos(7*pi/4 - 2*pi) since sin(x) = cos(270+x) </h3>
